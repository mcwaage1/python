```python
#random coin toss
import random

heads = 1
tails = 2
tosses = int(input("How many times do you want to toss the coin?"))
def main():
    for toss in range(tosses):
        if random.randint(heads, tails) == heads:
            print("Heads")
        else:
            print("Tails")
            
main()
```
