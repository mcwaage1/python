```python
# lottery number generator and winner verifier

import random

def main():
    
    ticket = str(input("Do you want to play the lottery?"))
    
    number = int(input("How many numbers are there on the ticket?"))
    
    num = int(input("What is the range of the numbers on the ticket?"))
        
    if ticket == "yes" or ticket == "y":
        
        lottery = []
        
        for x in range(number):
            
            numbers = random.randint(1, num)
            
            lottery.append(numbers)
        
        print("Here's your ticket", lottery)
        
    compare = str(input("Do you want to know if you won?"))
        
    if compare == "yes" or compare == "y":
        
        generate = []
        
        for x in range(number):
            
            numbers = random.randint(1, num)
            
            generate.append(numbers)
            
        print(generate)
        
        if generate == lottery:
            
            print("Wow! You actually won!")
            
        else:
            
            print("Sorry, you didn't win.")
            
            win = str(input("Do you want to know how many tickets it would take to win?"))
    
            if win == "y" or win == "yes":
            
                new_ticket = []
                
                count = 0

                while new_ticket != lottery:
                    
                    new_ticket = []
                    
                    for x in range(number):
                        
                        numbers = random.randint(1, num)

                        new_ticket.append(numbers)

                        count += 1
                        
                    print(new_ticket)

                else:
                    
                    print("It would have taken you", count, "tickets to win!")
                
            else:
                
                print("Ok, bye")
            
main()
```
