```python
# more string method modifications


def main():
    
    character = str(input("Enter a letter, number, or whatever."))
    
    num = int(input("How far do you want the character to go up?"))
    
    for count in range(0, num):
        
        print(character * count)
        
    for count in range(num - 1, 0, -1):
        
        print(character * count)
        
    for count in range(num - 1, 0, -1):
        
        print(character * count)
        
    for count in range(0, num):
        
        print(character * count)
    
main()
```
